### 데이터 정제

```python
# None : 파이썬에서 누락데이터 표시방식

a = np.array([1,2, None,4, 5])
a
```

```python
# a.sum() None값이 있으면 연산이 안됨
```

```python
# NaN : 누락된 수치 데이터

a = np.array([1,2,np.nan, 4, 5])
a.dtype
```

```python
0+ np.nan
```

```python
np.nan +np.nan
```

```python
a.sum(), a.min(), a.max()
```

```python
#nan값이 있을 경우의 함수

np.nansum(a), np.nanmin(a), np.nanmax(a)
```

```python
# Series에서는 None도 NaN으로 나타남

pd.Series([1,2, None, 4, np.nan])
```

```python
s = pd.Series(range(5), dtype=int)
s
```

```python
s[0] = None
s
```

```python
s[3] = np.nan
s
```

```python
#Null 값의 처리

s = pd.Series([1,2,np.nan, 'string', None])
s
```

```python
s.isnull()
```

```python
#NaN이 아닌 값만 보여줘

s[s.notnull()]
```

```python
#NaN인 값은 드랍해버려

s.dropna()
```

### 값 치환

```python
# 엉뚱한 값들의 치환

s = pd.Series([1,2,-999, 3, -1000, 4])
s
```

```python
s.replace(-999, np.nan)
```

```python
s.replace([-999,-1000], np.nan)
```

```python
#nan을 특정값으로 치환하기

s.replace([-999,-1000], ['요렇게도 되나',0])
```
