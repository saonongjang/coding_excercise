```python
#논리연산
data = {'class':['A','B','C','A','B','C','C'],
        'Name':['S1','S2','S3','S4','S5','S6','S7'],
        'Age':[20,19,21,22,24,25,26],
        'Score':[90.,95.,75.,80.,70.,85.,90.],
        }
df = pd.DataFrame(data)
df
```

```python
df.loc[df['Score'] >=80] #80점 이상인 사람만 추출
```

```python
df.loc[df['Score'] >=80, 'Name'] #80점 이상인 사람의 이름만 추출
```

```python
df.loc[df['Score'] >=80, ['Name', 'Age']] #80점 이상인 사람 이름과 나이를 추출
```

```python
#pass와 Fail항목을 추가

df['Result'] ='NONE' #default값으로 NONE

df.loc[ df['Score'] >=80, 'Result' ] = 'Pass'
df.loc[ df['Score'] <80, 'Result' ] = 'Fail'
df
```

```python
#index로 합격한 사람만 분류하기

idx = (df['Result'] =='Pass')
df.loc[idx]
```

```python
#합격한 사람들을 정렬

df.loc[ idx ].sort_values('Score')
```

```python
df.loc[ idx ].sort_values('Score', ascending=False)  #내림차순으로 정렬
```

```python
#추출한 데이터를 엑셀로 저장하기

data_sorted = df.loc[ idx ].sort_values('Score', ascending=False)
data_sorted.to_excel('data_sorted.xlsx', index = False) #index는 저장을 않음
```

```python
# 꺼꾸로 엑셀을 불러오기

df_import = pd.read_excel('data_sorted.xlsx')
df_import
```
