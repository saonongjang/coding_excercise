### GroupBy 연산


```python
df = pd.DataFrame({'c1':['a','a','b','b','c','d','b'],
                   'c2':['A','B','B','A','D','C','c'],
                   'c3':np.random.randint(7),
                   'c4' :np.random.random(7)})
df
```

```python
df.dtypes
```

```python
df['c3'].groupby(df['c1']).mean()
```

```python
df['c4'].groupby(df['c2']).std()
```

```python
df['c4'].groupby([df['c1'], df['c2']]).mean()
```

```python
df['c4'].groupby([df['c1'], df['c2']]).mean().unstack()
```

```python
df.groupby('c1').mean()
```

```python
df.groupby(['c1','c2']).size()
```

```python
for c1, group in df.groupby('c1'):
  print(c1)
  print(group)
```

```python
for (c1,c2), group in df.groupby(['c1','c2']):
  print((c1,c2))
  print(group)
```

```python
#특정열에 대한 mean값만 도출

df.groupby(['c1','c2'])['c4'].mean()
```

```python
df.groupby('c1')['c3'].count()
```

```python
df.groupby('c1')['c3'].median()
```

```python
df.groupby('c1')['c4'].std()
```

```python
#여러가지 결과를 한꺼번에 출력하기

df.groupby(['c1','c2'])['c4'].agg(['mean','min', 'max', 'std'])
```

```python
df.groupby(['c1','c2'], as_index=False)['c4'].mean() #c1,c2가 인덱스가 사라지고 칼럼으로 변환됨
```

```python
df.groupby(['c1','c2'], group_keys=False)['c4'].mean() #c1,c2가 인덱스가 사라지고 칼럼으로 변환됨
```

```python
def top(df,n=3,column='c1'):
  return df.sort_values(by=column)[-n:]

top(df,n=5)
```

```python
df.groupby('c1').apply(top)
```

### 피벗테이블

```python
df.pivot_table(['c3','c4'],
               index=['c1'],
               columns=['c2'])
```

```python
df.pivot_table(['c3','c4'],
               index=['c1'],
               columns=['c2'],
               margins=True)
```

```python
df.pivot_table(['c3','c4'],
               index=['c1'],
               columns=['c2'],
               margins=True,
               aggfunc=sum)
```

```python
df.pivot_table(['c3','c4'],
               index=['c1'],
               columns=['c2'],
               margins=True,
               aggfunc=sum,
               fill_value=0)
```
