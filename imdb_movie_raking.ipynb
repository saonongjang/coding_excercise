```python
#IMDB에서 영화 데이터를 불러오기 연습

#https://bit.ly/2NyxdAG
from bs4 import BeautifulSoup
import requests
import re

# 데이터 불러오기
url = "https://www.imdb.com/chart/top"
res = requests.get(url)
soup = BeautifulSoup(res.text, 'lxml')

#각 변수의 정의
movies = soup.select('td.titleColumn')
links = [a.attrs.get('href') for a in soup.select('td.titleColumn a')]
crew = [a.attrs.get('title') for a in soup.select('td.titleColumn a')]
ratings = [b.attrs.get('data-value') for b in soup.select('td.posterColumn span[name=ir]')]
votes = [b.attrs.get('data-value') for b in soup.select('td.ratingColumn strong')]

imdb = []

# 각데이터별로 딕셔너리 형태로 저장하기
for index in range(0, len(movies)):
    # Seperate movie into: 'ranking', 'title', 'year'
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    ranking = movie[:len(str(index))-(len(movie))]
    data = {"movie_title": movie_title,
            "year": year,
            "ranking": ranking,
            "star_cast": crew[index],
            "rating": ratings[index],
            "vote": votes[index],
            "link": links[index]}
    imdb.append(data)

for item in imdb:
    print(item['ranking'], '-', item['movie_title'], '('+item['year']+') -', 'Starring:', item['star_cast'], 'rating:', item['rating'])
  
```
