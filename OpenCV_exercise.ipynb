```python
#OpenCV 연습

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
#이미지 불러오기

im =cv.imread('민결1.jpg')

#BGR -->색깔의 순서가 RGB가 아니라 BGR이 디폴트
plt.figure()
plt.imshow(im)
plt.title('Original')
```

```python
#지나치게 파란색임으로 RGB로 바꾸기

rgb=cv.cvtColor(im, cv.COLOR_BGR2RGB)
plt.figure()
plt.imshow(rgb)
plt.title("RGB")
```

```python
#gray 이미지로 바꾸기

gray = cv.cvtColor(im,cv.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray, cmap='gray')
plt.title("Gray")
```

```python
#blur처리된 이미지와 함께 보기

blur = cv.blur(im,(50,50))  #숫자가 커질수록 심하게 blur처리됨
blur = cv.cvtColor(blur,cv.COLOR_BGR2RGB)
plt.subplot(121)
plt.imshow(rgb)
plt.title('RGB')
plt.subplot(122)
plt.imshow(blur)
plt.title("Blur")
```

```python
#Edge Detect

edges = cv.Canny(gray, 10,20)
plt.subplot(121)
plt.imshow(gray,cmap = 'gray')
plt.title('GRAY')
plt.subplot(122)
plt.imshow(edges, cmap = 'gray')
plt.title("Edges Detected")
```
