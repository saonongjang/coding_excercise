### 데이터의 결합
#### concate()와 append

```python
# 필요한 라이브러리 불러오기

import numpy as np
import pandas as pd
```

```python
s1 = pd.Series(['a', 'b'], index=[1,2])
s2 = pd.Series(['c', 'd'], index=[3,4])
pd.concat([s1,s2])
```

```python
# dataframe을 자동으로 만들어 주기 위한 함수 만들기

def create_df(cols,idx):
  data = {c:[str(c.lower()) + str(i) for i in idx] for c in cols}
  return pd.DataFrame(data,idx)
```

```python
df1 = create_df('AB', [1,2])
df1
```

```python
df2= create_df('AB', [3,4])
df2
```

```python
# df의 concate

pd.concat([df1,df2])
```

```python
df3 = create_df('AB', [0,1])
df3
```

```python
df4 = create_df('CD', [1,2])
df4
```

```python
# 가로 concate

pd.concat([df3,df4])
```

```python
# 동일한 인덱스를 가지고 있는 것을 합치는 방법

pd.concat([df3,df4],axis=1)
```

```python
# pd.concat([df1,df3], verify_integrity=True)  #합칠수 없는 df를 확인하는 방법
```

```python
# 그래도 억지로 df를 합치고 싶다면

pd.concat([df1,df3], ignore_index=True)
```

```python
# key 값을 주어서 합치는 방법

pd.concat([df1,df2], keys=['x','y'])
```

```python
df5 = create_df('ABC', [1,2])
df6 = create_df('BCD', [3,4])
pd.concat([df5,df6])

```

```python
# 중복되는 column의 데이터만 추출

pd.concat([df5,df6], join='inner')
```

```python
#df5에 df6를 추가

df5.append(df6)
```

#### 데이터의 병합과 조인

```python
#좀 더 큰 데이터프레임을 생성

df1 = pd.DataFrame({'학생':['홍길동','이순신','임꺽정','김유신'],
                    '학과':['경영학과','교육학과','컴퓨터학과', '통계학과']})
df1
```

```python
df2 = pd.DataFrame({'학생':['홍길동','이순신','임꺽정','김유신'],
                    '입학년도':['2000', '2001', '2002', '2003']})
df2
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
df4 = pd.DataFrame({'학과':['경영학과','교육학과','컴퓨터학과', '통계학과'],
                    '학과장':['황희', '장영실', '백민결', '최영']})
df4
```

```python
pd.merge(df3,df4)
```

```python
df5 = pd.DataFrame({'학과':['경영학과', '교육학과', '교육학과', '컴퓨터학과', '컴퓨터학과', '통계학과'],
                    '과목': ['경영개론', '기초수학', '물리학', '프로그래밍','운영체제', '확률론']})
df5
```

```python
pd.merge(df1,df5)
```

```python
pd.merge(df1,df2, on='학생')
```

```python
df6 = pd.DataFrame({'이름':['홍길동', '이순신', '임꺽정', '김유신'],
                    '성적':['A', 'B', 'C', 'A+']})
df6
```

```python
pd.merge(df1,df6, left_on ='학생', right_on = '이름' )
```

```python
pd.merge(df1,df6, left_on ='학생', right_on = '이름' ).drop('이름', axis=1)
```

```python
mdf2
```

```python
pd.merge(mdf1, mdf2, left_index = True, right_index=True)
```

```python
mdf1.join(mdf2)
```

```python
df7 =pd.DataFrame({'이름':['홍길동', '이순신', '임꺽정'], 
                   '주문음식' :['햄버거', '피자', '짜장면']})
df7
```

```python
df8 =pd.DataFrame({'이름':['홍길동', '이순신', '김유신'], 
                   '주문음료' :['콜라', '사이다', '커피']})
df8
```

```python
pd.merge(df7,df8)
```

```python
pd.merge(df7,df8, how='inner') #단순이 merge하는 것과 동일한 결과
```

```python
pd.merge(df7,df8, how='outer') #중복이 안되는 것도 보여줌
```

```python
pd.merge(df7,df8, how='left') # 좌측 df(df7)기준으로 보여줌
```

```python
pd.merge(df7,df8, how='right')
```
